```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("M16-Amazon-Challenge").config("spark.driver.extraClassPath","/content/postgresql-42.2.16.jar").getOrCreate()
```

### Load Amazon Data into Spark DataFrame

```python
from pyspark import SparkFiles
url = "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Video_Games_v1_00.tsv.gz"
spark.sparkContext.addFile(url)
df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get("amazon_reviews_us_Video_Games_v1_00.tsv.gz"), sep="\t", header=True, inferSchema=True)
df.show()
```

### Re-Create vine_df


```python
vine_df = df.select(['review_id', 'star_rating', 'helpful_votes', 'total_votes', 'vine', 'verified_purchase'])
vine_df.show()
```

### Filter vine_df for total_votes count >= 20

```python
df1 = vine_df.filter(vine_df.total_votes >=20)
df1.show(10)
```

### Filter vine_df_filtered to retrieve all rows where helpful_votes divided by total_votes >= 0.5

```python
df2 = df1.filter((df1.helpful_votes / df1.total_votes) >= 0.5)
df2.show(5)
```

```python
# Filter where a review was written as part of the Vine program (paid), vine == 'Y'
vine_program_df = df2.filter(df2.vine == 'Y')
vine_program_df.show(5)
```

```python
# Filter where the review was not part of the Vine program (unpaid), vine == 'N'
non_vine_program_df = df2.filter(df2.vine == 'N')
non_vine_program_df.show(5)
```

```python
vine_review_count = vine_program_df.count()
vine_review_count
```

```python
non_vine_review_count = non_vine_program_df.count()
non_vine_review_count
```

```python
# Determine the total number of reviews
total_reviews_df = vine_review_count + non_vine_review_count
total_reviews_df
```

```python
# The number of 5-star reviews (vine)
vine_program_five_stars = vine_program_df.filter(vine_program_df.star_rating == 5).count()
vine_program_five_stars
```

```python
# The number of 5-star reviews (non_vine)
non_vine_program_five_stars = non_vine_program_df.filter(non_vine_program_df.star_rating == 5).count()
non_vine_program_five_stars
```

```python
# Total number of 5-star reviews
total_five_star_reviews = vine_program_five_stars + non_vine_program_five_stars
total_five_star_reviews
```

```python
# Percentage of 5-star reviews for the two types of reviews (paid)
vine_five_star_percentage = (vine_program_five_stars / vine_review_count) * 100
vine_five_star_percentage
```

```python
# Percentage of 5-star reviews for the two types of reviews (unpaid)
non_vine_five_star_percentage = (non_vine_program_five_stars / non_vine_review_count) * 100
non_vine_five_star_percentage
```
